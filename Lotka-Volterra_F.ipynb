## 1. t = 0
## Descripción: Asigna el valor 0 a la variable t.
## Propósito: t representa el tiempo inicial en la simulación. Comienza en 0, lo cual es común en simulaciones donde se va a avanzar en el tiempo desde un punto de partida inicial.
## 2. x = 1.0
## Descripción: Asigna el valor 1.0 a la variable x.
## Propósito: x representa la población inicial de una especie o entidad en el modelo. En el contexto de un modelo como el de Lotka-Volterra (que estudia las dinámicas de depredador-presa), x podría ser la población inicial de presas.

## Propósito: y representa la población inicial de otra especie o entidad en el modelo. Si se trata de un modelo de depredador-presa, y podría ser la población inicial de depredadores.

```python
# initial time and populations
t = 0; x = 1.0; y = 0.5
```

## 1. t_list = []
## Descripción: Crea una lista vacía y la asigna a la variable t_list.
## Propósito: t_list almacenará los valores de tiempo (t) en cada paso de la simulación. A medida que avanza la simulación, los valores de t se agregarán a esta lista, permitiendo un registro completo de cómo evoluciona el tiempo durante la simulación.
## 2. x_list = []
## Descripción: Crea una lista vacía y la asigna a la variable x_list.
## Propósito: x_list almacenará los valores de la población x en cada paso de la simulación. Esto permite rastrear cómo cambia la población x (por ejemplo, la población de presas en un modelo de depredador-presa) con el tiempo.
## 3. y_list = []
## Descripción: Crea una lista vacía y la asigna a la variable y_list.
## Propósito: y_list almacenará los valores de la población y en cada paso de la simulación. Esto permitirá rastrear los cambios en la población y (por ejemplo, la población de depredadores) a lo largo del tiempo.

```python
# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []
```

## 1. t_list.append(t)
## Descripción: Añade el valor actual de t (que es 0) a la lista t_list.
## Propósito: Esto asegura que la lista t_list comience con el valor inicial del tiempo. En este caso, t_list ahora contendrá [0].
## 2. x_list.append(x)
## Descripción: Añade el valor actual de x (que es 1.0) a la lista x_list.
## Propósito: Esto inicializa la lista x_list con el valor de la población x en el tiempo inicial. Ahora, x_list contendrá [1.0].
## 3. y_list.append(y)
## Descripción: Añade el valor actual de y (que es 0.5) a la lista y_list.
## Propósito: Esto inicializa la lista y_list con el valor de la población y en el tiempo inicial. Ahora, y_list contendrá [0.5].

```python
# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)
```

## while t < max_time:
## Descripción: Este es el comienzo de un bucle while que continuará ejecutándose mientras t sea menor que max_time.
## Propósito: El bucle simula la evolución de las variables en el tiempo. Terminará cuando el tiempo t alcance o supere max_time (en este caso, 100).
## 2. t = t + dt
## Descripción: Incrementa el tiempo t por un pequeño intervalo dt.
## Propósito: Esto avanza el tiempo en la simulación. dt es el paso de tiempo y determina la resolución temporal de la simulación.
## 3. x = x + (a*x - b*x*y)*dt
## Descripción: Calcula la nueva población x utilizando una ecuación diferencial simple y luego actualiza el valor de x.
## Propósito: Esto modela el cambio en la población x durante el pequeño intervalo de tiempo dt. La ecuación (a*x - b*x*y) representa la tasa de cambio de x, que depende de los parámetros a (tasa de crecimiento de x) y b (efecto de la interacción con y), y de la población actual x y y. Multiplicando por dt se obtiene el cambio en x en ese pequeño intervalo de tiempo.
## 4. y = y + (-c*y + e*x*y)*dt
## Descripción: Calcula la nueva población y utilizando una ecuación diferencial simple y luego actualiza el valor de y.
## Propósito: Esto modela el cambio en la población y durante el intervalo de tiempo dt. La ecuación (-c*y + e*x*y) representa la tasa de cambio de y, donde -c*y es la tasa de decrecimiento natural de y, y e*x*y es la tasa de crecimiento de y debido a la interacción con x.
## 5. t_list.append(t)
## Descripción: Añade el nuevo valor de t a la lista t_list.
## Propósito: Registra el tiempo actual en la lista t_list para que puedas seguir la evolución temporal del sistema.
## 6. x_list.append(x)
## Descripción: Añade el nuevo valor de x a la lista x_list.
## Propósito: Registra la población actual x en la lista x_list para que puedas seguir la evolución de x a lo largo del tiempo.
## 7. y_list.append(y)
## Descripción: Añade el nuevo valor de y a la lista y_list.
## Propósito: Registra la población actual y en la lista y_list para que puedas seguir la evolución de y a lo largo del tiempo.

```python
while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
```

## 1. p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
## Descripción: Llama a la función plot de matplotlib.pyplot para trazar los datos.
## Propósito: Este comando crea una gráfica con dos curvas:
## t_list, x_list, 'r': Traza la población x en función del tiempo t usando una línea roja ('r').
## t_list, y_list, 'g': Traza la población y en función del tiempo t usando una línea verde ('g').
## Argumentos:
## t_list: Lista de valores de tiempo.
## x_list: Lista de valores de la población x.
## 'r': Especifica el color rojo para la línea que representa a x.
## y_list: Lista de valores de la población y.
## 'g': Especifica el color verde para la línea que representa a y.
## linewidth = 2: Establece el grosor de las líneas en 2 unidades, lo que hace que las líneas sean más gruesas y fáciles de ver.
## 2. plt.show()
## Descripción: Llama a la función show de matplotlib.pyplot para mostrar la gráfica.
## Propósito: Esta línea abre una ventana gráfica o renderiza el gráfico en un entorno como Jupyter Notebook. Sin esta línea, el gráfico no se mostraría visualmente.

```python
# Plot the results    
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
plt.show()
```

### PRIMER MODIFICACION DE ESTADO : La primera modificación aumentare la tasa de crecimiento de la población en presas y este se vera reflejado en la tasa a los que los depredadores aumentan al consumir presas como se mostrara en la siguiente grafica.


```python
a = 5; b = 5; c = 0.3;  e = 0.2
dt = 0.001; max_time = 100

# initial time and populations
t = 0; x = 1.0; y = 0.5

# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []

# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)

while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Plot the results    
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
plt.show()
```

### Vemos un claro aumenta en la grafica con altos picos, esto indica que las presas aumentaron y los depredadores también.
