```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los vértices del triángulo original
triangle = np.array([[0, 0], [1, 0], [0.5, 1], [0, 0]])  # Agregamos el último vértice para cerrar el triángulo

def plot_triangle(ax, triangle, title, color='blue'):
    """Función para dibujar el triángulo"""
    ax.plot(triangle[:, 0], triangle[:, 1], color=color, lw=2)
    ax.fill(triangle[:, 0], triangle[:, 1], color=color, alpha=0.3)
    ax.set_title(title)
    ax.set_xlim([-2, 3])
    ax.set_ylim([-2, 3])
    ax.grid(True)
    ax.set_aspect('equal')

# Funciones de transformación
def translate(triangle, dx, dy):
    """Traslación del triángulo por (dx, dy)"""
    translation_matrix = np.array([dx, dy])
    return triangle + translation_matrix

def scale(triangle, sx, sy):
    """Escalar el triángulo por (sx, sy)"""
    scaling_matrix = np.array([[sx, 0], [0, sy]])
    return np.dot(triangle, scaling_matrix)

def rotate(triangle, angle_deg):
    """Rotar el triángulo por un ángulo en grados"""
    angle_rad = np.radians(angle_deg)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)], 
                                [np.sin(angle_rad), np.cos(angle_rad)]])
    return np.dot(triangle, rotation_matrix)

# Crear una figura con subplots para mostrar las transformaciones
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Triángulo original
plot_triangle(axs[0, 0], triangle, "Original")

# Traslación
translated_triangle = translate(triangle, dx=1.5, dy=1)
plot_triangle(axs[0, 1], translated_triangle, "Traslacion", color='green')

# Escalación
scaled_triangle = scale(triangle, sx=2, sy=1.5)
plot_triangle(axs[1, 0], scaled_triangle, "Escalacion", color='red')

# Rotación
rotated_triangle = rotate(triangle, angle_deg=45)
plot_triangle(axs[1, 1], rotated_triangle, "Rotacion (45°)", color='purple')

# Mostrar la figura
plt.tight_layout()
plt.show()
```
