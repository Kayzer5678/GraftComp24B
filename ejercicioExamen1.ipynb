```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vertices = np.array([[0, 0], [2, 3], [5, 3], [6, 1], [8, 2], [7, 5], [10, 7], [5, 8], [3, 6], [1, 7], [0, 5], [0, 0]])
fig, ax = plt.subplots()

for i in range(len(vertices)-1):
    x_values = [vertices[i][0], vertices[i+1][0]]
    y_values = [vertices[i][1], vertices[i+1][1]]
    ax.plot(x_values, y_values, 'b-', marker='o', markerfacecolor='red') 

ax.set_xlim(-1, 11)
ax.set_ylim(-1, 11)

plt.show()
```
